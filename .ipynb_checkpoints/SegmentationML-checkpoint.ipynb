## Data Cleaning
<hr>

```python
dfseg.isnull().sum()
```

__Comment__ : it seems there are missing values in Minimum_Payments and Credit_limit,

We will use one of preprocessing technique by filling the NaN value with Mean().
reason : The feature that has missing value is numerical data, beside that this dataset contain only 8950 data which is quite few so it's better not to remove the data

```python
# Changing NaN Values of Minimum_Payments and Credit_limit to mean

dfseg['MINIMUM_PAYMENTS'] = dfseg['MINIMUM_PAYMENTS'].fillna(int(dfseg['MINIMUM_PAYMENTS'].mean()))
dfseg['CREDIT_LIMIT'] = dfseg['CREDIT_LIMIT'].fillna(int(dfseg['CREDIT_LIMIT'].mean()))
```

```python
dfseg.isnull().sum()
```

### Identifying Outliers in Numerical Data

```python
j = 1
plt.figure(figsize=(18,18))

for i in dfseg.columns[1:] :
    plt.subplot(6,3,j)
    plt.boxplot(dfseg[i])
    plt.title(f'Outlier in {i}')
    j += 1
    
plt.show()
```

```python
Q1 = dfseg[dfseg.columns[1:]].quantile(0.25)
Q3 = dfseg[dfseg.columns[1:]].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
dfseg_new =dfseg[~(( dfseg[dfseg.columns[1:]] < (Q1 - 1.5 * IQR)) |( dfseg[dfseg.columns[1:]] > (Q3 + 1.5 * IQR))).any(axis=1)].copy()
print(dfseg.shape)
print(dfseg_new.shape)
```

```python
j = 1
plt.figure(figsize=(18,18))

for i in dfseg_new.columns[1:] :
    plt.subplot(6,3,j)
    plt.boxplot(dfseg_new[i])
    plt.title(f'Outlier in {i}')
    j += 1
    
plt.show()
```

__Comment__ : it seems that removing the outlier will reduce the data greatly(approximately aroung 5000 data will be dropped). even after that when we try to plot the new data it still have tremendous amount of outlier. 

__Solution__: since we will use K-Means Algorithm to cluster this data and for K-means to perform effectively, we have to scale the data. We will use Robust Scaler that are robust to outliers.

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaler.fit(dfseg[dfseg.columns[1:]])
```

```python
rst = scaler.transform(dfseg[dfseg.columns[1:]])
```

```python
dffeaturex=pd.DataFrame()
j = 0
for i in dfseg.columns[1:]:
    dffeaturex[i] =  rst[:, j]
    j += 1
```

```python
dffeaturex.head()
```

### Elbow Method (Defining amount of cluster)

```python
from sklearn.cluster import KMeans
```

```python
k_range = np.arange(1,50)
inertia = []
for i in k_range:
    model = KMeans(n_clusters= i)
    model.fit(dffeaturex)
    inertia.append(model.inertia_)

plt.plot(k_range, inertia)
plt.ylabel('Sum Squared Error (inertia)')
plt.xlabel('k Values')
plt.show()
```

```python
plt.plot(k_range, inertia)
plt.plot(k_range[10], inertia[10],'ro')
plt.ylabel('Sum Squared Error (inertia)')
plt.xlabel('k Values')
plt.show()
```

__Comment__ : From the above result, it seem that from this dataset we could interpret that the amount of cluster is between 8-12

in this case i choose 10 as the amount of cluster

```python
MLKM =KMeans(n_clusters = 10)
```

```python
MLKM.fit(dffeaturex)
```

```python
yp = MLKM.predict(dffeaturex)
```

```python
dfseg['ClusterPred'] = yp
dfseg.head()
```

### Cluster Evaluation Metrics (Without Label)

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score,davies_bouldin_score
```

__1. Silhouette Coefficient__

```python
silhouette_score(dffeaturex,yp)
```

__2. Calinski-Harabasz Index__

```python
calinski_harabasz_score(dffeaturex,yp)
```

__3. Davies-Bouldin Index__

```python
davies_bouldin_score(dffeaturex,yp)
```
