### Defining Target

Late payments can have a significant effect on credit card score.The late payment is added to credit report when your payment is more than 30 days late. An entry is added to your credit report and can stay for seven years. 

in this case :
- Not Overdue: if status below 30 days due (Status : 0, C, X) (Label : 0)
- Overdue : if stats above 30 days due ( Status : 1,2,3,4,5) (Label : 1)


```python
dfcr['target'] = dfcr['STATUS'].apply(lambda x : 0 if x == 'X' or x == 'C' or x == '0' else 1)
```

```python
dfcr = dfcr.groupby('ID').max()
dfcr = dfcr.drop(['MONTHS_BALANCE','STATUS'], axis=1)
```

```python
# Merge Dataset

dfapproval = dfcr.merge(dfapp, left_on='ID', right_on='ID')
dfapproval.head()

# Renaming Columns
dfapproval.rename(columns={
    'CODE_GENDER':'Gender',
    'FLAG_OWN_CAR':'Car',
    'FLAG_OWN_REALTY':'Property',
    'CNT_CHILDREN':'NoChild',
    'AMT_INCOME_TOTAL':'Income',
    'NAME_EDUCATION_TYPE':'Education',
    'NAME_FAMILY_STATUS':'MartialStat',
    'NAME_HOUSING_TYPE':'HousingType',
    'FLAG_EMAIL':'Email',
    'NAME_INCOME_TYPE':'IncomeType',
    'FLAG_MOBIL':'MobilePhone',
    'FLAG_WORK_PHONE':'Workphone',
    'FLAG_PHONE':'Phone',
    'CNT_FAM_MEMBERS':'FamSize',
    'OCCUPATION_TYPE':'Occupation'},inplace=True)

dfapproval.head()
```

### Binary Data Visualization

```python
def BinViz (notes,df,feature, color,fc):
    y1 = df[feature][(df[feature]== 1)].count()
    y2 = df[feature][(df[feature]== 0)].count()
    data = [y1,y2]
    w,l,p = plt.pie(data, labels = notes, colors = color,
                 autopct = lambda x: '{:.1f}% \n ({})'.format(x, int(round(x * sum(data)/100))))
    for i in p:
        i.set_color(fc)

def BinVizTarget (notes,df,feature,value, color,fc):
    y1 = df[feature][(df[feature]== value) & (df['target'] == 1) ].count()
    y2 = df[feature][(df[feature]== value) & (df['target'] == 0) ].count()
    data = [y1,y2]
    w,l,p = plt.pie(data, labels = notes, colors = color,
                 autopct = lambda x: '{:.1f}% \n ({})'.format(x, int(round(x * sum(data)/100))))
    for i in p:
        i.set_color(fc)
```

```python
plt.figure('Binary Visualization', figsize=(18,12))

plt.subplot(231)
BinViz(['Male','Female'], dfapproval_new,'Gender', ['#00008B','#B22222'],'w' )
plt.title('%Gender Overall')

plt.subplot(232)
BinViz(['Have Car', 'No Car'], dfapproval_new,'Car', ['#228B22','#B22222'],'w' )
plt.title('%Car Overall')

plt.subplot(233)
BinViz(['Have Property', 'No Property'], dfapproval_new,'Property', ['#228B22','#B22222'],'w' )
plt.title('%Property Overall')

plt.subplot(234)
BinViz(['Have Phone','No Phone'], dfapproval_new,'Phone', ['#228B22','#B22222'],'w' )
plt.title('%Phone Overall')

plt.subplot(235)
BinViz(['Have Workphone', 'No Workphone'], dfapproval_new,'Workphone', ['#228B22','#B22222'],'w' )
plt.title('%Workphone Overall')

plt.subplot(236)
BinViz(['Have Email', 'No Email'], dfapproval_new,'Email', ['#228B22','#B22222'],'w' )
plt.title('% Email Overall')

plt.suptitle('Overall Visualization')

plt.savefig('static/hometemplate/images/fig0.jpg')
plt.show()
```

### Feature Importance

```python
# Dummies for Categorical Data
categorical = ['IncomeType','Education','MartialStat','HousingType','Occupation']
dfdummy = pd.get_dummies(dfapproval_new[categorical])
```

```python
dffeaturex = pd.concat([dfapproval_new[['Gender','Car','Property','NoChild','Income','Age','LOE','Workphone','Phone','Email','FamSize']],dfdummy],
                      axis='columns')
dffeaturey = dfapproval_new['target']
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(dffeaturex,dffeaturey)
feat_importances = pd.Series(model.feature_importances_, index=dffeaturex.columns)
feat_importances.nlargest(10).plot(kind='barh')

plt.savefig('static/hometemplate/images/fig9.jpg')
plt.show()
```

### Select K best

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(dffeaturex,dffeaturey)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(dffeaturex.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

## Data Cleaning
<hr>

```python
dfseg.isnull().sum()
```

__Comment__ : it seems there are missing values in Minimum_Payments and Credit_limit,

We will use one of preprocessing technique by filling the NaN value with Mean().
reason : The feature that has missing value is numerical data, beside that this dataset contain only 8950 data which is quite few so it's better not to remove the data

```python
# Changing NaN Values of Minimum_Payments and Credit_limit to mean

dfseg['MINIMUM_PAYMENTS'] = dfseg['MINIMUM_PAYMENTS'].fillna(int(dfseg['MINIMUM_PAYMENTS'].mean()))
dfseg['CREDIT_LIMIT'] = dfseg['CREDIT_LIMIT'].fillna(int(dfseg['CREDIT_LIMIT'].mean()))
```

```python
dfseg.isnull().sum()
```

### Identifying Outliers in Numerical Data

```python
j = 1
plt.figure(figsize=(18,18))

for i in dfseg.columns[1:] :
    plt.subplot(6,3,j)
    plt.boxplot(dfseg[i])
    plt.title(f'Outlier in {i}')
    j += 1
    
plt.show()
```

```python
Q1 = dfseg[dfseg.columns[1:]].quantile(0.25)
Q3 = dfseg[dfseg.columns[1:]].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
dfseg_new =dfseg[~(( dfseg[dfseg.columns[1:]] < (Q1 - 1.5 * IQR)) |( dfseg[dfseg.columns[1:]] > (Q3 + 1.5 * IQR))).any(axis=1)].copy()
print(dfseg.shape)
print(dfseg_new.shape)
```

```python
j = 1
plt.figure(figsize=(18,18))

for i in dfseg_new.columns[1:] :
    plt.subplot(6,3,j)
    plt.boxplot(dfseg_new[i])
    plt.title(f'Outlier in {i}')
    j += 1
    
plt.show()
```
